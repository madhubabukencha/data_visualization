```python
%matplotlib inline
import seaborn as sns
```

```python
# Loading tips data set
tips = sns.load_dataset('tips')
tips.sample(7)
```

### seaborn.lmplot()

Plot data and regression model fits across a FacetGrid. <br>

This function combines regplot() and FacetGrid. It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.<a href="https://seaborn.pydata.org/generated/seaborn.lmplot.html"> Read More </a>

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'],
           scatter_kws={'s':70} # specifying size of points on the grap)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time', row='sex')
```
