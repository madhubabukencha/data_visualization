### To install Seaborn

```python
# Importing Seaborn
import seaborn as sns
```

```python
# Loading tips data set
tips = sns.load_dataset('tips')
tips
```

### sns.displot()
This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots. <a hreaf="https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot">Read More</a>


```python
sns.displot(tips["total_bill"], 
            bins=20 # bins means nothing but number bar to display
           )
```

```python
sns.displot(tips["total_bill"], kind='kde')
```

```python
sns.displot(tips["total_bill"], kde=True)
```

### sns.jointplot()

Draw a plot of two variables with bivariate and univariate graphs. <br>

This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly. <a href="https://seaborn.pydata.org/generated/seaborn.jointplot.html"> Read More </a>

```python
sns.jointplot(data=tips, x="total_bill", y="tip")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", 
              kind="reg", height=6, 
              ratio=3, # bar heights outside the graph)
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", hue="sex", kind="kde")
```

### sns.pairplot()

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.<a href="https://seaborn.pydata.org/generated/seaborn.pairplot.html"> Read More</a>

```python
sns.pairplot(tips, hue="sex", palette="rocket")
```

```python
sns.pairplot(tips, hue="sex", diag_kind="hist", palette="viridis", markers=["o", "s"])
```

### seaborn.rugplot
Plot marginal distributions by drawing ticks along the x and y axes.

This function is intended to complement other plots by showing the location of individual observations in an unobstrusive way.

```python
sns.rugplot(data=tips, x="tip")
```
