```python
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
# Loading datasets
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

### Correlation (r)
Finding relationship between 2 or more than variables <br>
Example: <br>
If pertrol usage increase along with it petrol prices increase: positive correlation (r<=1) <br>
pertrol usage decreased(variable1) but petrol price increases(variable2): negative correlation(-1 <= r) <br>
pertrol usage use increase(variable1) and children stopped playing(variable2): No correlation between 2 variables(r=0)<br>
<b>Source</b>:<a href="https://youtu.be/lBB4stn3exM">Correlation Coefficient</a>

```python
my_data = {'usage':[4, 5, 6, 4], 'price':[40, 50, 60, 1]}
df = pd.DataFrame(my_data)
df.head()
x = df.corr()
x
```

```python
sns.heatmap(x, annot=True)
```

```python
# To make perfect co-relation(r=1) changing fourth element
my_data = {'usage':[4, 5, 6, 4], 'price':[40, 50, 60, 40]}
df = pd.DataFrame(my_data)
x = df.corr()
sns.heatmap(x, annot=True)
```

```python
# To see negative co-relation
my_data = {'usage':[40, 50, 60, 40], 'price':[10, 5, 1, 10]}
df = pd.DataFrame(my_data)
x = df.corr()
sns.heatmap(x, annot=True)
```

```python
# To see negative non co-relation
my_data = {'usage':[40, 50, 60, 40], 'price':[184, 1, 100, 10]}
df = pd.DataFrame(my_data)
x = df.corr()
sns.heatmap(x, annot=True)
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm")
```

### Lets see how many passengers travelled each month

```python
fc = flights.pivot_table(index="month", columns="year", values="passengers")
fc
```

```python
sns.heatmap(fc, cmap="magma", linecolor='white', linewidth=1)
```

### Seaborn.clustermap()

Plot a matrix dataset as a hierarchically-clustered heatmap.

```python
flights = sns.load_dataset('flights')
fc = flights.pivot_table(index="month", columns="year", values="passengers")
fc
```

```python
sns.clustermap(fc)
```
