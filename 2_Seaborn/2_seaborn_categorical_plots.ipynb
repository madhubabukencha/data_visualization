Plots are basically used for visualizing the relationship between variables. Those variables can be either be completely numerical or a category like a group, class or division.

```python
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
# Loading data
tips = sns.load_dataset("tips")
tips.head(3)
```

### Seaborn.barplot()

A barplot is basically used to aggregate the categorical data according to some methods and by default itâ€™s the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.

```python
# By default it will show the average total bill  by male and female
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
# To find standard deviation
# In statistics, the standard deviation is a measure of the amount of variation or dispersion of a set of values.
# A low standard deviation indicates that the values tend to be close to the mean of the set
# while a high standard deviation indicates that the values are spread out over a wider range.
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

### seaborn.countplot()

countplot is used to draw the graph by counting the values

```python
# We are plot below number
# tips["sex"].value_counts()
sns.countplot(x="sex", data=tips)
```

### seaborn.boxplot()
Draw a box plot to show distributions with respect to categories. <a href="https://seaborn.pydata.org/generated/seaborn.boxplot.html">Read more</a>

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue="sex")
```

### seaborn.violinplot()
Draw a combination of boxplot and kernel density estimate. <a href="https://seaborn.pydata.org/generated/seaborn.violinplot.html">Read more</a>

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue="smoker")
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue="smoker", split=True, palette="rocket")
```

### seaborn.stripplot()
Draw a scatterplot where one variable is categorical.

A strip plot can be drawn on its own, but it is also a good complement
to a box or violin plot in cases where you want to show all observations 
along with some representation of the underlying distribution. <a href="https://seaborn.pydata.org/generated/seaborn.stripplot.html"> Read more</a>

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue="smoker")
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue="sex", jitter=True, palette="rocket")
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue="smoker", split=True, palette="Set2")
sns.stripplot(x='day', y='total_bill', data=tips, color="black", jitter=True)
```

### seaborn.factorplot()

```python
sns.catplot(x='day', y='total_bill', data=tips, kind="bar",  palette="mako")
```
