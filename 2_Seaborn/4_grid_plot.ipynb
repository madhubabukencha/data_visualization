```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading iris dataset
iris = sns.load_dataset('iris')
iris.sample(7)
```

```python
# Finding unique species
iris['species'].unique()
```

### seaborn.PairGrid()

It is a Subplot grid for plotting pairwise relationship in a dataset. This object maps each variable (column) in a dataset onto a column and row in a grid of multiple axes.

```python
# It is similar to pairplot. Use PairGrid when you need more flexibility.
# sns.pairplot(iris, hue="species")
```

```python
g = sns.PairGrid(iris, palette="rocket")
g.map(plt.scatter)
```

```python
# To change diagonal, upper and lower styles
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.add_legend()
```

### seaborn.FacetGrid()
Multi-plot grid for plotting conditional relationships

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')
```
