In this program we will find a <b>sinθ</b> value of different angles and using them we draw a graph

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np


angles = [0, 30, 45, 60, 90, 180, 270, 360]
values = np.sin(angles)
# 'bo-' means blue color, round points, solid lines
plt.plot(angles, values, "ro-")

# zip joins x and y coordinates in pairs
for x,y in zip(angles,values):
    z = "{:.2f}".format(y)
    plt.annotate((f"sin{x}={z}"), # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,5), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

#Adding additional feature to the figure
plt.title("sinθ values")
plt.xlabel("Angles")
plt.ylabel("Values")
plt.grid(True)
plt.rcParams["figure.figsize"] = (13, 7)
plt.show()
```
