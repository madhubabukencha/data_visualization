```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_values = np.linspace(1, 10, num=10, dtype=int)
print(x_values)
y_values = x_values**2
print(y_values)
```

### plt.plot() method

A simple method to generate graphs

```python
plt.plot(x_values, y_values,'*-', color="green", label="Square")
plt.plot(x_values, x_values**3,'--', color="red", label="Cubes")
# 1: upper right corner, 2: upper left corner, 3: lower left corner, 4: lower right corner
# 0: let matplotlib decide the optimal location
plt.legend(loc=2)
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.title("Square Chart")
```

### plt.subplot() method
To draw multiple plots in single graph

```python
# The figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1)
plt.plot(x_values, y_values, "r")
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1, 2, 2)
plt.plot(x_values, y_values+5, "b")
```

### plt.figure() method

```python
# This module is used to control the default spacing of the subplots
# and top level container for all plot elements
fig = plt.figure()

# Syntax: add_axes(rect, projection=None, polar=False, **kwargs)
# rect: The dimensions [left, bottom, width, height] of the new Axes.
# you can understand it better when you are working with multiple plots
axes = fig.add_axes([0.9, 0.9, 0.7, 0.5], 
                    xlabel="Numbers",
                    ylabel="Squares", 
                    title="Square Table")

axes.plot(x_values, y_values, "r")
```

```python
fig = plt.figure()
# Root or outside plot
axes_1 = fig.add_axes([0.1, 0.1, 1, 0.9])

# Squares Plot
axes_2 = fig.add_axes([0.3, 0.6, 0.3, 0.3])
axes_2.plot(x_values, y_values, "--", color="red")
axes_2.set_title("Squares")
axes_2.set_xlabel("Number")
axes_2.set_ylabel("Numbers Square")

# Logs square plot
axes_3 = fig.add_axes([0.6, 0.2, 0.4, 0.3])
axes_3.set_title("log square values")
axes_3.plot(x_values, np.log(y_values))
```

### plt.subplots()

This method helps in creating multiple graphs. With subplots we no need to create add_axes. Those are created by default

```python
fig, axes = plt.subplots(nrows=3,
                         ncols=3)
axes[0][0].plot(x_values, y_values)
axes[0, 0].set_title("Square")
axes[0, 0].set_xlabel("Number")
axes[0, 0].set_ylabel("Numbers Square")

axes[0][1].bar(y_values, x_values)
axes[0, 1].set_title("With Bar")

axes[0][2].scatter(y_values, x_values, marker=">")
# Provide proper spacing between graphs
plt.tight_layout()
```

### figsize size & DPI
dot per inch(DPI): The higher the dpi the better the resolution and the better copy/print quality

```python
fig = plt.figure(figsize=(4, # width
                          3, # height
                         ))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x_values, y_values)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 3))
axes[0].plot(x_values, y_values)
axes[0].grid(True)
axes[1].plot(y_values, x_values)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.7)
```

#### To save a graph

```python
fig.savefig("subplots.png", dpi=500)
```

### Adding legend to graph

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x_values, x_values**2, label="Square")
ax.plot(x_values, x_values**3, label="Cubes")
# Here you can find other location values
# https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.legend.html
ax.legend(loc = 0)
```

### Plot appearance

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x_values, y_values,
          color="#417D7A", # you can also specify RGB values ex 
          linewidth=3, # or lw is for specifying line width
          alpha=0.5, # Indicates the line transperency
          linestyle="--", # or ls To specify line style
          marker = "o", # To add marker to the line,
          markersize = 15, # To specify marker size,
          markerfacecolor = "yellow", # Color of the marker
          markeredgewidth= 3,
          markeredgecolor="red")
# To set x and y axis limits or to perticular portion of the graph
axes.set_xlim([0.5, 2.5])
axes.set_ylim([0, 5])
```
