### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Importing Required Packages</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Data Prepartion</span>

```python
x = np.linspace(1, 10, num=10, dtype="int")
x
```

### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Different Kind of Plots</span>

#### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Scatter Plot</span>

```python
plt.scatter(x=x, y=x**2)
```

```python
sizes = [5, 3, 1, 1]
explode = (0, 0.1, 0, 0)
plt.pie(sizes, 
        shadow=True,
        autopct='%1.1f%%',
        labels=["English", "Maths", "Telugu", "Physics"],
        labeldistance=1.1,
        explode = explode,
        colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'],)
plt.show()
```
